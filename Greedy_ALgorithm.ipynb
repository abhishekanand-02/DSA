```python
class Item:
    def __init__(self, profit, weight):
        self.profit = profit
        self.weight = weight

## Function definition
def maxResult(arr, M):
    # sort by profit/weight ratio (descending)
    arr.sort(key=lambda x: x.profit / x.weight, reverse=True)
    
    maxProfit = 0.0
    for item in arr:
        if item.weight <= M:
            M -= item.weight
            maxProfit += item.profit
        else:
            maxProfit += item.profit * (M / item.weight)
            break
    return maxProfit

## Driver Code
arr = [
    Item(25, 5),
    Item(75, 10),
    Item(100, 12),
    Item(50, 4),
    Item(45, 7),
    Item(90, 9),
    Item(30, 3)
]

M = 37
result = maxResult(arr, M)
print("Max profit from the greedy Algorithm:", result)
```
