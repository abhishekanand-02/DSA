## Depth First search Traversal

```python
## function definition
def DFT(visited, graph, node):
    if node not in visited:
        print(node, end = " ")
        visited.add(node)

        for adjacentNode in graph[node]:
            if adjacentNode not in visited:
                DFT(visited, graph , adjacentNode)
        



# Driver code 
## storing the data of the graph using adjancency list
graph = {
    '5':['3','7'],
    '3':['2','4'],
    '7':['8'],
    '2': [],
    '4':['8'],
    '8': []
}
visited = set()

DFT(visited, graph, '5')
    
```

## Breadth first Search 

```python
from collections import deque
queue = deque()

def BST(visited, graph, node):
    visited.add(node)
    queue.appendleft(node)

    while queue:
        node = queue.pop()
        print(node, end = " ")
        for adjancentNodes in graph[node]:
            if adjancentNodes not in visited:
                visited.add(adjancentNodes)
                queue.appendleft(adjancentNodes)


## Driver code 
visited = set()

graph = {
    'A':['B','C','D'],
    'B':['E'],
    'C':['E','F'],
    'D': ['F'],
    'E':['G'],
    'F':['G'],
    'G': []
}


BST(visited, graph , 'A')
```
